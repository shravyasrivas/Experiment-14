```python
import pandas as pd
import numpy as np
data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price': [55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold': [25, 60, 35, 80, 50, 15]
}
df=pd.DataFrame(data)
print(df)
```

```python
#dividing these prices into categories is called data binning
#Data Binning
bins1=[0,10000,30000,60000]
labels1=['Low ','Medium ','High ']
df['Price_Category']=pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)
```

```python
#now convert the unit sold into categorical data
bins2=[0,30,60,100]
labels2=['Low','Medium','High']
df['Sales_Category']=pd.cut(df['Units_Sold'], bins=bins2, labels=labels2)
print(df)
```

```python
#Data Formatting
#Convert Data Types int to flt
print(df.dtypes)
df['Units_Sold']= df['Units_Sold'].astype(float)
print(df.dtypes) #this command should be run again and again as it rewrites the data type
```

```python
df #in fractions
```

```python
#Coverting case to uppercase
df['Product']=df['Product'].str.upper()
print(df)
#.title is for making it a title eg:the tiger .title will give The Tiger
```

```python
#Rounding values: rounds of a demical to 2 decimal values .round(2)
df['Price']=df['Price'].round(2)
print(df)
```

```python
#Additional Operations
#Sorting: ascending order
sorted_df= df.sort_values(by='Price')
print(sorted_df)
#.implace 2 would combine 2 tables and write them in the same memory location
```

```python
#sorting in descending order
sorted_df=df.sort_values(by='Price',ascending=False)
print(sorted_df)
```

```python
#Display Unique Values
print(df['Price_Category'].unique())
```

```python
#New Dataset: Food Delivery Orders
import pandas as pd
import numpy as np

data = {
'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8'],
'Order_Value':[250,450,120,600,300,150,500,220],
'Delivery_Time':[30,45,25,60,35,20,50,28],
'Distance_km':[5,8,3,10,6,2,9,4]
}

df = pd.DataFrame(data)

print(df)
```

```python
#binning order value
bins=[0,200,400,700]
labels=['Low Order','Medium Order','High Order']
df['Order_Category']=pd.cut(df['Order_Value'],bins=bins,labels=labels)
print(df)
```

```python
#binning delivery time
bins=[0,20,40,60]
labels=['Fast ','Medium ','Slow ']
df['Delivery_Time_Category']=pd.cut(df['Delivery_Time'],bins=bins,labels=labels)
print(df)
```

```python
#binning distance
bins=[0,4,8,12]
labels=['Near','Medium','Far']
df['Distance_km_Category']=pd.cut(df['Distance_km'],bins=bins,labels=labels)
print(df)
```

```python
#Data Formatting
#Convert Data Types int to flt
print(df.dtypes)
df['Distance_km']= df['Distance_km'].astype(float)
print(df.dtypes) #this command should be run again and again as it rewrites the data type
```

```python
#Coverting case to uppercase
df['Delivery_Time_Category']=df['Delivery_Time_Category'].str.upper()
print(df)
```

```python
#Rounding values: rounds of a demical to 2 decimal values .round(2)
df['Order_Value']=df['Order_Value'].round(2)
print(df)
```

```python
#Additional Operations
#Sorting: ascending order
sorted_df= df.sort_values(by='Order_Value')
print(sorted_df)
```

```python
#sorting in descending order
sorted_df=df.sort_values(by='Order_Value',ascending=False)
print(sorted_df)
```
